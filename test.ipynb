```python
import os 

folder = '/data/kinetics400/valid_subset1'

files_dict = {}
# base_path 아래의 모든 디렉토리에 대해 반복
for i, (root, dirs, files) in enumerate(os.walk(folder)):
    if i>=1:
        label = os.path.basename(root)
        matched_files = [file for file in files if any(file.endswith(ext) for ext in [".mp4"])]
        if matched_files:
            files_dict[label] = matched_files
            
print(files_dict)
```

```python
print(len(files_dict))
```

```python
total_files = sum(len(files) for files in files_dict.values())
print(total_files)
```
